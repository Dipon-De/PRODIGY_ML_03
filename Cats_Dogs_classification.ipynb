```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pickle
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from tensorflow.keras.applications import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input
```

```python
dir = "D:/Programming/ML/Prodigy_Infotech/cats_dogs/dataset"
categories = ['cats', 'dogs']
data = []
```

```python
def load_images_and_augment(dir, categories):
    data = []
    for category in categories:
        path = os.path.join(dir, category)
        label = categories.index(category)
        for img in os.listdir(path):
            imgpath = os.path.join(path, img)
            pet_img = cv2.imread(imgpath, 1)  # Read in color
            try:
                pet_img = cv2.resize(pet_img, (128, 128))
                data.append([pet_img, label])
                # Data Augmentation
                data.append([cv2.flip(pet_img, 1), label])  # Horizontal flip
            except Exception as e:
                pass
    return data
```

```python
data = load_images_and_augment(dir, categories)
```

```python
def extract_features(data):
    model = VGG16(weights='imagenet', include_top=False, input_shape=(128, 128, 3))
    features = []
    labels = []
    for image, label in data:
        image = np.expand_dims(image, axis=0)
        image = preprocess_input(image)
        feature = model.predict(image)
        feature = feature.flatten()
        features.append(feature)
        labels.append(label)
    return np.array(features), np.array(labels)
```

```python
features, labels = extract_features(data)
```

```python
scaler = StandardScaler()
features = scaler.fit_transform(features)
```

```python
pca = PCA(n_components=200)
features_pca = pca.fit_transform(features)
```

```python
x_train, x_test, y_train, y_test = train_test_split(features_pca, labels, test_size=0.20)
```

```python
param_grid = {'C': [0.1, 1, 10], 'gamma': ['scale', 'auto'], 'kernel': ['rbf']}
svc = SVC()
clf = GridSearchCV(svc, param_grid, refit=True, verbose=2)
clf.fit(x_train, y_train)
```

```python
print(f"Best Parameters: {clf.best_params_}")
best_model = clf.best_estimator_
predictions = best_model.predict(x_test)
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy: {accuracy}")
```
